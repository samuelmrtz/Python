```python
import soccerdata as sd

# 1. Le decimos que queremos usar FBref para la Premier League de la temporada actual
# Puedes cambiar "ENG-Premier League" por "ESP-La Liga", "GER-Bundesliga", etc.
# Puedes usar '2324' para la temporada pasada o '2425' para la actual
fbref = sd.FBref(leagues="ENG-Premier League", seasons="2425") 

print("📡 Conectando a FBref y descargando la base de datos...")

# 2. Descargamos el calendario completo de esa temporada
df_calendario = fbref.read_schedule()

# 3. Filtramos para ver solo los partidos que YA se jugaron (que ya tienen resultado)
df_jugados = df_calendario[df_calendario['score'].notna()].copy()

# 4. FBref entrega la columna 'score' (Goles Reales) y la columna 'xg' (Goles Esperados).
# Vamos a mostrar los últimos 5 partidos jugados para que veas qué datos trae
print("✅ Datos descargados con éxito. Mostrando los últimos 5 partidos:")
display(df_jugados[['date', 'home_team', 'away_team', 'score', 'xg_home', 'xg_away']].tail(5))

# 5. (Opcional) Ver los partidos que FALTAN por jugarse (La próxima jornada)
df_por_jugar = df_calendario[df_calendario['score'].isna()].copy()
print("\n📅 Próximos partidos programados (Jornada a predecir):")
display(df_por_jugar[['date', 'home_team', 'away_team']].head(5))
```
